```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv("iris.csv")
data
```

```python
features=['PetalLengthCm','PetalWidthCm']
x=data[features]
```

```python
x_standardized=StandardScaler().fit_transform(x)
```

```python
dbscan=DBSCAN(eps=0.5,min_samples=5)
clusters=dbscan.fit_predict(x_standardized)
clusters
```

```python
centroids=pd.DataFrame(x_standardized).groupby(clusters).mean().values
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.scatter(x['PetalLengthCm'],x['PetalWidthCm'],c=clusters,cmap='viridis',edgecolor='k')
plt.title('DBSCAN Clustering of Species')
plt.xlabel('PetalLengthCm')
plt.ylabel('PetalWidthCm')
plt.show()
```
